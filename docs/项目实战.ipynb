## 原生TCP服务
直接使用TcpListener::bind()创建并绑定监听器， bind()方法返回一个io::Result<TcpListener>，如果绑定成功，返回一个TcpListener实例，否则返回一个错误。
```rust
use std::net::TcpListener;
let listener = TcpListener::bind("xx:xx").unwrap();
```
thread::spawn 签名如下：
```rust
pub fn spawn<F, T>(f: F) -> JoinHandle<T>
where
    F: FnOnce() -> T,
    F: Send + 'static,
    T: Send + 'static,
```
spawn 选择FnOnce作为F闭包的特征约束，原因是闭包作为任务只需被线程执行一次即可
F的Send特征约束是为了将相关参数从一个线程传递到另一个线程。
‘static生命周期约束，是因为我们并不知道线程需要多久时间来执行任务
