# 类型转换
rust中内置了一些基本类型之间的转换
## as转换
```rust
let a : u32 = 10;
let b : u64 = 20;
if a < (b as u32) {
    println!("success");
}
```
我们一般将范围较小的转换成范围较大的， 因为较大的转换为较小的可能发生截断，造成与预期不符的结果

# newtype和类型别名
## newtype
newtype 就是使用元组结构体的方式将已有类型包裹起来
```rust
struct Meters(u32)
```
newtype的使用场景：
* 隐藏内部类型细节
* 对于某些场景，只有newtype可以很好的解决
* 自定义类型可以让我们给出更具意义和可读性的类型名。

上述说的某些场景，就是绕过孤儿规则(要为类型A实现特征T， 那么A或者T必须至少有一个在当前的作用范围内)的场景
```rust
use std::fmt;
struct Wrapper(Vec<String>);
impl fmt::Display for Wrapper {
    fn fmt(&self, &mut fmt::Formatter) ->fmt::Result {
        write!(f, "[{}]", self.0.join(", "))
    }
}

fn main() {
    let w = Wrapper(vec![String::from("hello"), String::from("world")]);
    println!("w={}",w);
}

```

## 类型别名
类型别名：
```rust
type Meters = u32
```
类型别名不是一个独立的全新的类型，而是某一个类型的别名
```rust
type Meters = u32;
let x: u32 = 5;
let y : Meters = 6;
println!("x+y = {}", x+y);
//如上代码可正常运行
```
* 类型别名仅仅是别名，只是为了让可读性更好，并不是全新类型，newtype才是
* 类型别名无法实现为外部类型实现外部特征等功能，而newtype可以

**类型别名可以减少模版代码的使用**
